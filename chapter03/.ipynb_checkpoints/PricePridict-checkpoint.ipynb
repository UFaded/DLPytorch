```python
features = pd.concat((train_data.iloc[:, 1:-1], test_data.iloc[:, 1:]))#把id舍去
print(features.shape)
```

```python
#定义形式如下
#df = pd.DataFrame(np.random.randn(4, 3), columns=list('bde'), index=['utah', 'ohio', 'texas', 'oregon'])
#所以可以顺利的取index

numeric_features = features.dtypes[features.dtypes != 'object'].index

#本行的打印结果：
#Index(['MSSubClass', 'LotFrontage', ...]

#-------
#定义形式如下
#df = pd.DataFrame(np.random.randn(4, 3), columns=list('bde'), index=['utah', 'ohio', 'texas', 'oregon'])
#所以可以顺利的取index
#-------

#不是object的那些列，有int64等数值型，因此它说要对连续数值的特征做标准化。公式如下。
#另：默认是dim=0为列，所以就是取某x列的均值，和方差。

features[numeric_features] = features[numeric_features].apply(
    lambda x: (x - x.mean()) / (x.std()))

# 标准化后，每个数值特征的均值变为0，所以可以直接用0来替换缺失值
features[numeric_features] = features[numeric_features].fillna(0)

all_features = pd.get_dummies(features, dummy_na=True)
print(all_features.shape) # (2919, 331)
```

```python
train_size = train_data.shape[0]
train_features = torch.tensor(all_features[:train_size].values, dtype=torch.float)
test_features = torch.tensor(all_features[train_size:].values, dtype=torch.float)
train_labels = torch.tensor(train_data.SalePrice.values, dtype=torch.float).view(-1, 1)

hidden_layer = 124
```

```python
def get_net(num_input=331):
    net = nn.Sequential(
        nn.Linear(num_input, 1),
        #加上这个的话，过拟合了大哥！
        # nn.ReLU(),
        # nn.Linear(hidden_layer, 1)
    )
    for param in net.parameters():
        nn.init.normal_(param, mean=0, std=0.01)
    return net
```

```python
loss_func = nn.MSELoss()
#对数均方根误差是个什么神仙计算
def log_rmse(net, features, labels):
    with torch.no_grad():
        # 将小于1的值设成1，使得取对数时数值更稳定
        # 反正最终的clipped_preds >= 1的
        clipped_preds = torch.max(net(Variable(features)), torch.tensor(1.0))
        rmse = torch.sqrt(2 * loss_func(clipped_preds.log(), labels.log()).mean())
    return rmse.item() #计算后返回标量的item
```

```python
def get_k_fold_data(k, i, X, y): #训练k次
    assert k > 1
    #取整数的亚子
    fold_size = X.shape[0] // k #分成k组 ??有疑问 float->integer
    #注意注意！因为报错：需要两个tensor才能合并。所以建一个空的tensor
    x_train, y_train = torch.Tensor(), torch.Tensor() 
    for j in range(k):
        index = slice(j * fold_size, (j + 1)*fold_size) #截取片段
        x_part, y_part = X[index, :], y[index] #用于与上一个即x_train or y_train进行拼接
        if j == i:
            x_valid, y_valid = x_part, y_part #把第i个片段作为验证数据
        elif x_train is None: #防止只分了一个片段
            x_train, y_valid = x_part, y_part
        else:
            print()
            x_train = torch.cat((x_train, x_part), dim=0)
            y_train = torch.cat((y_train, y_part), dim=0)
    return x_train, y_train, x_valid, y_valid
```

```python
def train(net, train_features, train_labels, test_features, test_labels,
          num_epochs, learning_rate, weight_decay, batch_size):
    torch_dataset = Data.TensorDataset(train_features, train_labels)
    train_loader = Data.DataLoader(torch_dataset, batch_size=batch_size, shuffle=True)
    optimizer = optim.Adam(params=net.parameters(), lr=learning_rate, weight_decay=weight_decay)
    train_ls, test_ls = [], []
    for epoch in range(num_epochs):
        for x, y in train_loader:
            x = Variable(x)
            out = net(x)
            loss = loss_func(out.float(), y.float())
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

        train_ls.append(log_rmse(net, train_features, train_labels))
        if test_labels is not None:
            test_ls.append(log_rmse(net, test_features, test_labels))
    return train_ls, test_ls
```

```python
def k_fold(k, x_train, y_train, num_epochs, learning_rate, weight_decay, batch_size):
    train_l_sum, valid_l_sum = 0., 0.
    for i in range(k):
        #玄学... C++想都没想过 还能这么整
        data = get_k_fold_data(k, i, x_train, y_train)
        net = get_net(x_train.shape[1])
        train_ls, valid_ls = train(net, *data, num_epochs, learning_rate, weight_decay, batch_size)
        #只要知道最后一行 为最终误差
        train_l_sum += train_ls[-1]
        valid_l_sum += valid_ls[-1]

        if i == 0:
            d2l.semilogy(range(1, num_epochs + 1), train_ls, 'epochs', 'rmse',
                         range(1, num_epochs + 1), valid_ls,
                         ['train', 'valid'])
        print('fold %d, train rmse %f, valid rmse %f' % (i, train_ls[-1], valid_ls[-1]))
    return train_l_sum / k, valid_l_sum / k
```

```python
k, num_epochs, lr, weight_decay, batch_size = 5, 100, 5, 0, 64
train_l, valid_l = k_fold(k, train_features, train_labels, num_epochs, lr, weight_decay, batch_size)
print('%d-fold validation: avg train rmse %f, avg valid rmse %f' % (k, train_l, valid_l))
```

```python
def train_and_pred(train_features, test_features, train_labels, test_data,
                   num_epochs, lr, weight_decay, batch_size):
    net = get_net(train_features.shape[1])
    train_ls, _ = train(net, train_features, train_labels, None, None, num_epochs, lr, weight_decay, batch_size)
    #画图
    d2l.semilogy(range(1, num_epochs + 1), train_ls, 'epochs', 'rmse')
    print('train rmse %f' % train_ls[-1])
    preds = net(test_features).detach().numpy()
    test_data['SalePrice'] = pd.Series(preds.reshape(1, -1)[0])
    submission = pd.concat([test_data['Id'], test_data['SalePrice']], axis=1)
    submission.to_csv('./submission.csv', index=False)

train_and_pred(train_features, test_features, train_labels, test_data, num_epochs, lr, weight_decay, batch_size)
```

**思路**：

1. 使用pd从csv中读取数据，

2. 进行基本的数据预处理：

   本次学习到的方法：把训练数据和测试数据一起处理；将非object的数据标准化；用0替代缺失值；将可能有多种选择数据的列进行拓展；转化为tensor，为忘了设置dtype=float。

3. 搭建net网络（在一个项目中，如果确定了网络的层数，最好把输入层的层数设置为需要后来传入的变量num_input，就像本节展示的那样。可能这里不太实用，但是代码结构会比较好，自己定义一个实在是太low，一点也不高手！）。

4. 搭建好网络后，最好也将训练代码段写成一个函数。

   实现的功能有：数据类型的转换tensor->dataset；小批量数据loader；优化函数Adam；

   循环内部大体工作是一样的，但是因为传入参数test_label可能是None的，所以在计算误差rmse的时候，需要提前判断。

   之所以可能为None，是因为最后画图的时候，只会计算训练集的数据误差，不再考虑测试集了，此时loss是仅仅只由训练集那100多个组成的，所以传入的测试集是None。

   该函数之后在K折的时候会反复调用，单独检验的时候也是，写函数避免代码冗杂。

5. 计算log_rmse和loss_func是不一样的，前者是用来评价模型的，后者是平方损失函数，优化模型的。

6. K折交叉过程：把所有训练集传入，分成已知的K份，依次取1～K作为测试集。

   在get_k_fold_data函数中，返回的是已经分好的训练集和测试集。思想就是，来到第 i 组的时候归测试集，其余的与训练集cat一下。torch.cat((a,b), dim=0)

   然后再把返回的train和test数据集都进行训练，并对test集进行模型评估！

   k_fold函数将训练的损失评估综合取平均。

   这样就完成了一个epoch的训练了！

7. 有一点注意：optimizer的优化是依赖loss_func来更新参数，然后log_rmse与loss_func相关连，所以把它作为损失函数也是很ok的。
